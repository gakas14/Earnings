## Data Cleaning - Check for Missing Values and Duplicates

Find and remove any duplicate rows.

```python
data = df_data.drop_duplicates()
#data.isna().values.any()
```

## Descriptive Statistics

```python
data.describe()
```

## Visualise the Features

```python
fig = px.scatter(data,
                x=data['S'],
                y=data['EARNINGS']
                )
fig.show()
#plt.scatter(data['S'], data['EARNINGS'])
```

# Split Training & Test Dataset

X_train, y_train = 80% of the dataset.
X_test, y_test = 20% of the dataset.

X_train, X_test = are the features(The features are all the other columns you'll use to predict the earnings.)
y_train, y_test = target (the earnings columns)

```python
target = data['EARNINGS']
features = data.drop('EARNINGS', axis=1)
features = features.replace(np.nan,0)
X_train, X_test, y_train, y_test = train_test_split(features,
                                                   target,
                                                   test_size=0.2,
                                                   random_state=10
                                                   )
```

## Simple Linear Regression

Only use the years of schooling to predict earnings. Use sklearn to run the regression on the training dataset. How high is the r-squared for the regression on the training data? 

```python
# Explanatory Variable
X = pd.DataFrame(X_train, columns=['S'])

#Target
y = pd.DataFrame(y_train)

simple_reg = LinearRegression()
simple_reg.fit(X,y)

#Determined the r-squared
rsquared = simple_reg.score(X,y)
print(f'The R-squared is: {rsquared:.2}')
```

#### the r-squared is very low

### Evaluate the Coefficients of the Model

Here we do a sense check on our regression coefficients. The first thing to look for is if the coefficients have the expected sign (positive or negative). 

Interpret the regression. How many extra dollars can one expect to earn for an additional year of schooling?

```python
simple_reg_coef = pd.DataFrame(data=simple_reg.coef_, columns=['Coefficient'])
simple_reg_coef
#simple_reg.intercept_[0]
```

```python
year =1
extra_dollars = year*simple_reg_coef + simple_reg.intercept_[0]
dollars = extra_dollars.values[0,0]
print(f'The extra dollars one can expect to earn for an additional year of schooling are: ${dollars:.3}')
```

### Analyse the Estimated Values & Regression Residuals

How good our regression is also depends on the residuals - the difference between the model's predictions ( 𝑦̂ 𝑖 ) and the true values ( 𝑦𝑖 ) inside y_train. Do you see any patterns in the distribution of the residuals?

```python
predicted_vals = simple_reg.predict(X)
residuals = (y - predicted_vals)
```

```python
# Original Regression of Actual vs. Predicted Earnings
plt.figure(dpi=100)
plt.scatter(x=y, y=predicted_vals, c='indigo', alpha=0.6)
plt.plot(y, y, color='cyan')
plt.title(f'Actual Earnings vs Predicted Earnings: $y _i$ vs $\hat y_i$', fontsize=17)
plt.xlabel('Actual Earnings $y _i$', fontsize=14)
plt.ylabel('Predicted Earnings $\hat y _i$', fontsize=14)
plt.show()

# Residuals vs Predicted Earnings
plt.figure(dpi=100)
plt.scatter(x=predicted_vals, y=residuals, c='indigo', alpha=0.6)
plt.title('Residuals vs Predicted Earnings', fontsize=17)
plt.xlabel('Predicted Earnings $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.show()
```

# Multivariable Regression

Now use both years of schooling and the years work experience to predict earnings. How high is the r-squared for the regression on the training data? 

```python
X= X_train[['S','EXP']]

y= y_train

multi_reg = LinearRegression()
multi_reg.fit(X,y)

rq= multi_reg.score(X,y)
print(f'The r-squared for the multivariable regression is: {rq:.3}')
```

### Evaluate the Coefficients of the Model

```python
reg_coef = pd.DataFrame(data=multi_reg.coef_, columns=['Coefficient'])
reg_coef
```

### Analyse the Estimated Values & Regression Residuals

```python
predicted_vals = multi_reg.predict(X)
residuals = (y - predicted_vals)
```

```python
# Original Regression of Actual vs. Predicted Earnings
plt.figure(dpi=100)
plt.scatter(x=y, y=predicted_vals, c='indigo', alpha=0.6)
plt.plot(y, y, color='cyan')
plt.title(f'Actual Earnings vs Predicted Earnings: $y _i$ vs $\hat y_i$', fontsize=17)
plt.xlabel('Actual Earnings $y _i$', fontsize=14)
plt.ylabel('Predicted Earnings $\hat y _i$', fontsize=14)
plt.show()

# Residuals vs Predicted Earnings
plt.figure(dpi=100)
plt.scatter(x=predicted_vals, y=residuals, c='indigo', alpha=0.6)
plt.title('Residuals vs Predicted Earnings', fontsize=17)
plt.xlabel('Predicted Earnings $\hat y _i$', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.show()
```

# Use Your Model to Make a Prediction

How much can someone with a bachelors degree (12 + 4) years of schooling and 5 years work experience expect to earn in 2011?

```python
#Starting Point: Average Values in the Dataset
average_vals = X.mean().values
expect_earn = pd.DataFrame(data=average_vals.reshape(1,len(X.columns)),
                          columns=X.columns)

expect_earn
```

```python
#Predict the earnings
expect_earn['S'] = 16
expect_earn['EXP'] = 5

predict_value = multi_reg.predict(expect_earn)
predict_value = predict_value[0]
print(f'Someone with a bachelors degree (12 + 4) years of schooling and 5 years work experience can expect to earn ${predict_value:.4} by hour in 2011')
```
